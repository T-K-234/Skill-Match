```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('users.db')

tables = ['users', 'skills', 'certificates', 'jobs', 'applications']

for table in tables:
    print(f"--- {table.upper()} ---")
    df = pd.read_sql_query(f"SELECT * FROM {table}", conn)
    print(df, "\n")

conn.close()
```
